# 04 — HMG Dataset Construction (Synthetic-to-Real Pipeline)

**Primary Dataset:** HMG (Hinglish-Medical-Grounding)  
**Target:** ~4,000 triplets → {Open-i X-ray, English Report, Synthetic Hinglish Query}  
**Method:** Llama-3-8B-Instruct + MMCQSD linguistic templates

---

## Pipeline Steps
1. Load preprocessed Open-i reports (formal English findings)
2. Load MMCQSD examples as linguistic templates for realistic code-switching
3. Build few-shot prompts for Hinglish query synthesis
4. Generate synthetic Hinglish queries using Llama-3-8B
5. Validate and quality-check generated queries
6. Compute CMI scores and bucket into low / medium / high
7. Assemble final HMG triplet dataset
